### Crescimento Populacional

**1. Crescimento Discreto (Geométrico):**

Usado para populações com gerações não sobrepostas (ex.: insetos, plantas anuais).

Fórmula: $N_{t+1} = N_t * \lambda$, onde $λ$ é a taxa de crescimento.

**2. Crescimento Contínuo (Exponencial):**

Usado para populações com reprodução contínua (ex.: humanos, bactérias).

Fórmula: $N(t) = N_0 * e^{rt}$, onde $r$ é a taxa intrínseca de crescimento.

**3. Crescimento Logístico (Limitação por Capacidade de Suporte):**

Incorpora um limite máximo ($K$) devido a recursos finitos.

Fórmula: $N(t)= \frac{K}{1 + \left(\frac{K - N_0}{N_0}\right) * e^{-rt}}$.

```python
# --- Parâmetros Iniciais ---
N0 = 100          # População inicial
lambda_ = 1.05     # Taxa de crescimento discreto (λ)
r = 0.05           # Taxa de crescimento contínuo (r)
rl = 0.1           # Taxa de crescimento contínuo no logístico
K = 1000          # Capacidade de suporte (logístico)
tempo = np.arange(0, 60, 1)  # 60 anos (discreto) ou unidades de tempo (contínuo)
```

```python
# --- 1. Crescimento Discreto (Geométrico) ---
pop_discreta = np.zeros_like(tempo, dtype=float)
pop_discreta[0] = N0
for t in range(1, len(tempo)):
    pop_discreta[t] = pop_discreta[t-1] * lambda_
```

```python
# --- 2. Crescimento Contínuo (Exponencial) ---
pop_continua = N0 * np.exp(r * tempo)
```

```python
# --- 3. Crescimento Logístico ---
pop_logistica = (K) / (1 + ((K - N0) / N0) * np.exp(-rl * tempo))
```

```python

plt.figure(figsize=(10, 5))
plt.plot(tempo, pop_discreta, 'b-o', label='Crescimento Discreto (λ=1.2)')
plt.plot(tempo, pop_continua, 'r--', label='Crescimento Contínuo (r=0.1)')
plt.plot(tempo, pop_logistica, 'g-', label=f'Crescimento Logístico (K={K})')
plt.axhline(y=K, color='k', linestyle=':', label='Capacidade de Suporte (K)')
plt.title("Modelos de Crescimento Populacional")
plt.xlabel("Tempo (anos)")
plt.ylabel("Tamanho da População (N)")
plt.legend()
plt.grid(True)
plt.show()
```

### Matriz de Leslie

A Matriz de Leslie para três classes etárias:

$L = 
\begin{vmatrix} 
    F_0 & F_1 & F_2 \\
    S_0 & 0   & 0   \\
    0   & S_1 & 0   \\
\end{vmatrix}$, 
onde $F$ é fertilidade na classe etária e $S$ é a sobrevivência na classe.

**Crescimento Populacional Discreto**: $N_{t+1} = N_t * \lambda$ => $N_{t+1} = N_t * L$

$
\begin{vmatrix}
    N_{0, t+1} \\
    N_{1, t+1} \\
    N_{2, t+1} \\
\end{vmatrix}$ = 
$\begin{vmatrix} 
    F_0 & F_1 & F_2 \\
    S_0 & 0   & 0   \\
    0   & S_1 & 0   \\
\end{vmatrix}$
$\times$
$\begin{vmatrix}
    N_{0, t} \\
    N_{1, t} \\
    N_{2, t} \\
\end{vmatrix}
$

```python
# --- Parâmetros Biológicos (ex.: Onça-pintada) ---
# Classes etárias: [filhotes, juvenis, adultos]
s = np.array([0.95, 0.8])      # Sobrevivência (filhotes->juvenis, juvenis->adultos)
f = np.array([0.0, 0.5, 0.7]) # Fertilidade (filhotes, juvenis, adultos)
```

```python
# --- Matriz de Leslie ---
L = np.array([
    [f[0], f[1], f[2]],  # Linha 1: fertilidade
    [s[0], 0.0,  0.0],  # Linha 2: sobrevivência filhotes -> juvenis
    [0.0,  s[1], 0.0]   # Linha 3: sobrevivência juvenis -> adultos
])
```

```python
# --- Vetor Inicial de População ---
N0 = np.array([30, 20, 10])  # [filhotes, juvenis, adultos]
```

```python
# --- Projeção no Tempo (20 anos) ---
tempo = 20
N = np.zeros((3, tempo))
N[:, 0] = N0

for t in range(1, tempo):
    N[:, t] = L @ N[:, t-1]  # Multiplicação matriz-vetor
```

```python
plt.figure(figsize=(10, 5))
plt.plot(N.T, '-o')  # Transposta para plotar classes por cor
plt.title("Dinâmica Populacional por Classe Etária (Matriz de Leslie)")
plt.xlabel("Anos")
plt.ylabel("Número de Indivíduos")
plt.legend(["Filhotes", "Juvenis", "Adultos"])
plt.grid(True)
plt.show()
```

```python
# --- Taxa de Crescimento (Autovalor Dominante) ---
autovalores, _ = np.linalg.eig(L)
lambda1 = np.max(autovalores)
print(f"Taxa de crescimento populacional (λ): {np.real(lambda1):.2f}")
```

```python
# --- Definir estrutura da Matriz de Leslie ---
# 3 classes etárias: [juvenis, adultos jovens, adultos]
def construir_leslie(s1, s2, f2, f3):
    """
    s1: sobrevivência de juvenis para adultos jovens
    s2: sobrevivência de adultos jovens para adultos
    f2: fertilidade de adultos jovens
    f3: fertilidade de adultos
    """
    L = np.array([
        [0.0, f2,   f3],    # Linha 1: fertilidade
        [s1,  0.0,  0.0],   # Linha 2: sobrevivência juvenis -> adultos jovens
        [0.0, s2,   0.0]    # Linha 3: sobrevivência adultos jovens -> adultos
    ])
    return L
```

```python
# --- Grade de parâmetros para análise ---
s1_vals = np.linspace(0.1, 0.9, 50)    # Sobrevivência juvenil
f3_vals = np.linspace(0.1, 2.0, 50)    # Fertilidade adulta
s2 = 0.8   # Sobrevivência adultos jovens fixa
f2 = 0.5   # Fertilidade adultos jovens fixa
```

```python
# --- Matriz para armazenar autovalores dominantes ---
autovalores = np.zeros((len(s1_vals), len(f3_vals)))
```

```python
# --- Calcular autovalor para cada combinação ---
for i, s1 in enumerate(s1_vals):
    for j, f3 in enumerate(f3_vals):
        L = construir_leslie(s1, s2, f2, f3)
        eigvals = np.linalg.eigvals(L)
        lambda_dominante = np.max(np.real(eigvals))
        autovalores[i, j] = lambda_dominante
```

```python
# --- Análise de limiar ---
print("=== ANÁLISE DE LIMIARES ===")
print(f"Para s2={s2} e f2={f2} fixos:")

# Encontrar combinações críticas
limiar_crescimento = autovalores > 1
percentual_crescimento = np.mean(limiar_crescimento) * 100

print(f"Percentual de combinações com λ > 1: {percentual_crescimento:.1f}%")

# Encontrar valor mínimo de f3 para λ > 1 com s1 máximo
s1_max = s1_vals[-1]
f3_limiar = f3_vals[np.where(autovalores[-1, :] > 1)[0][0]]
print(f"Com s1={s1_max:.2f}, necessário f3 > {f3_limiar:.2f} para crescimento")

# Encontrar valor mínimo de s1 para λ > 1 com f3 máximo
f3_max = f3_vals[-1]
s1_limiar = s1_vals[np.where(autovalores[:, -1] > 1)[0][0]]
print(f"Com f3={f3_max:.2f}, necessário s1 > {s1_limiar:.2f} para crescimento")
```

```python
import seaborn as sns

# --- Definir Estados de Qualidade Ambiental ---
estados = ['Bom', 'Regular', 'Crítico']
n_estados = len(estados)

# --- Matriz de Transição (probabilidades de transição entre estados) ---
# Linha i: estado atual | Coluna j: próximo estado
P = np.array([
    [0.7, 0.2, 0.1],   # Se BOM: 70% chance de permanecer, 20% p/ Regular, 10% p/ Crítico
    [0.3, 0.5, 0.2],   # Se REGULAR: 30% p/ Bom, 50% permanecer, 20% p/ Crítico
    [0.1, 0.3, 0.6]    # Se CRÍTICO: 10% p/ Bom, 30% p/ Regular, 60% permanecer
])
```

```python
# Verificar se as linhas somam 1 (propriedade de matriz estocástica)
print("Verificação da matriz estocástica (soma das linhas = 1):")
print(np.sum(P, axis=1))
```

```python
# --- Simulação da Cadeia de Markov ---
def simulacao_markov(P, estado_inicial, passos):
    """
    Simula uma cadeia de Markov
    
    Parâmetros:
    P: matriz de transição (n x n)
    estado_inicial: índice do estado inicial
    passos: número de passos da simulação
    """
    n_estados = P.shape[0]
    trajetoria = np.zeros(passos, dtype=int)
    trajetoria[0] = estado_inicial
    
    for t in range(1, passos):
        # Probabilidades de transição do estado atual
        prob_transicao = P[trajetoria[t-1], :]
        # Escolher próximo estado baseado nas probabilidades
        trajetoria[t] = np.random.choice(n_estados, p=prob_transicao)
    
    return trajetoria
```

```python
# --- Parâmetros da Simulação ---
estado_inicial = 0  # Começa no estado "Bom"
n_passos = 100      # 100 unidades de tempo (ex.: meses)
n_simulacoes = 10    # Número de trajetórias simuladas
```

```python
# --- Múltiplas Simulações ---
plt.figure(figsize=(12, 6))
for i in range(n_simulacoes):
    trajetoria = simulacao_markov(P, estado_inicial, n_passos)
    plt.plot(trajetoria, 'o-', alpha=0.7, label=f'Simulação {i+1}')

plt.yticks([0, 1, 2], estados)
plt.title('Simulações de Cadeia de Markov: Transição entre Estados de Qualidade Ambiental')
plt.xlabel('Tempo (passos)')
plt.ylabel('Estado de Qualidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# --- Distribuição Estacionária (longo prazo) ---
def calcular_distribuicao_estacionaria(P):
    """Calcula a distribuição estacionária resolvendo πP = π"""
    n = P.shape[0]
    # Resolver o sistema linear: π(P - I) = 0, sujeito a Σπ = 1
    A = np.vstack([(P.T - np.eye(n))[:-1], np.ones(n)])
    b = np.zeros(n)
    b[-1] = 1
    pi = np.linalg.lstsq(A, b, rcond=None)[0]
    return pi
```

```python
pi_estacionaria = calcular_distribuicao_estacionaria(P)
print("\nDistribuição Estacionária (longo prazo):")
for i, estado in enumerate(estados):
    print(f"{estado}: {pi_estacionaria[i]:.3f} ({pi_estacionaria[i]*100:.1f}%)")
```

```python
# --- Simulação de Monte Carlo para verificar distribuição estacionária ---
n_passos_longo = 10000
trajetoria_longa = simulacao_markov(P, estado_inicial, n_passos_longo)
```

```python
# Contar frequência de cada estado
frequencias = np.bincount(trajetoria_longa) / n_passos_longo
print("\nFrequências empíricas (simulação longa):")
for i, estado in enumerate(estados):
    print(f"{estado}: {frequencias[i]:.3f} ({frequencias[i]*100:.1f}%)")
```

```python
# --- Visualização da Distribuição Estacionária ---
plt.figure(figsize=(10, 5))
x_pos = np.arange(len(estados))
plt.bar(x_pos - 0.1, pi_estacionaria, width=0.4, label='Teórica', alpha=0.8)
plt.bar(x_pos + 0.1, frequencias, width=0.4, label='Empírica', alpha=0.8)
plt.xticks(x_pos, estados)
plt.title('Distribuição Estacionária: Teórica vs Empírica')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# --- Análise de Sensibilidade: Efeito de Poluição ---
print("\n=== ANÁLISE DE SENSIBILIDADE ===")

# Cenário 1: Aumento da poluição (mais transições para estados piores)
P_poluicao = np.array([
    [0.6, 0.3, 0.1],   # Mais chance de degradação
    [0.2, 0.5, 0.3],   # Mais chance de piorar
    [0.05, 0.25, 0.7]  # Mais difícil de recuperar
])

pi_poluicao = calcular_distribuicao_estacionaria(P_poluicao)
print("\nCom aumento de poluição:")
for i, estado in enumerate(estados):
    print(f"{estado}: {pi_poluicao[i]:.3f}")

# Cenário 2: Medidas de conservação (mais transições para estados melhores)
P_conservacao = np.array([
    [0.8, 0.15, 0.05],  # Mais estável no estado bom
    [0.4, 0.5, 0.1],    # Mais chance de melhorar
    [0.2, 0.4, 0.4]     # Mais fácil de recuperar
])

pi_conservacao = calcular_distribuicao_estacionaria(P_conservacao)
print("\nCom medidas de conservação:")
for i, estado in enumerate(estados):
    print(f"{estado}: {pi_conservacao[i]:.3f}")
```

```python
# --- Visualização Comparativa ---
plt.figure(figsize=(12, 5))
x_pos = np.arange(len(estados))
width = 0.25

plt.bar(x_pos - width, pi_estacionaria, width, label='Cenário Base', alpha=0.8)
plt.bar(x_pos, pi_poluicao, width, label='Com Poluição', alpha=0.8)
plt.bar(x_pos + width, pi_conservacao, width, label='Com Conservação', alpha=0.8)

plt.xticks(x_pos, estados)
plt.title('Comparação de Cenários: Efeito na Distribuição Estacionária')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
